## Interactive Plot with Angle as Variable

```python
def update_plot(angle, norm):
    
    theta=np.radians(angle)
    x=np.array([[1,0,0]])*norm
    y=np.array([[np.cos(theta), np.sin(theta), 0]])*norm
    z=np.cross(x, y)
    
#     print('x =',x)
#     print('y =',y)
#     print('z =',z)
    
    ax=plt.figure().add_subplot(projection='3d')
    ax.quiver(0, 0, 0, x[0][0], x[0][1], 0, color='blue')
    ax.quiver(0, 0, 0, y[0][0], y[0][1], 0, color='red')
    ax.quiver(0, 0, 0, 0, 0, z[0][2], color='purple')
    ax.set(xlim=[-4,4], ylim=[-4,4], zlim=[0,16])
    plt.show()

angle=widgets.IntSlider(min=0, max=540, step=15, value=0, description='Angle')
norm=widgets.IntSlider(min=1, max=5, step=1, value=4, description='Size')

interactive(update_plot, angle=angle, norm=norm)
```
